```python
# Sử dụng thư viện Ultralytics

#Sử dụng RT-DETR để phát hiện đối tượng
#Mô hình RT-DETR đã được tích hợp vào ultralytics,

from ultralytics import RTDETR

# Load mô hình RT-DETR pre-trained
model = RTDETR("rtdetr-l.pt")

# Dự đoán trên một ảnh
results = model("anhsuutam_74.jpg")

import cv2

# Vẽ bounding box lên ảnh gốc
annotated_img = results[0].plot()

# Hiển thị ảnh
cv2.imshow("Detected Image", annotated_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# hiển thị video theo frame
import cv2
from ultralytics import RTDETR

# Load mô hình RT-DETR
#model = RTDETR("rtdetr-l.pt")
model =  RTDETR("./Models_1_RTDETR_dataBienSoNhieuLoaiv4_datanoscale_anhmau1nhan/runs/detect/train/weights/best.pt")

# Mở video
#video_path = "video_test1.mp4"  # Đường dẫn video
video_path = './predict_data/video2.mp4'
cap = cv2.VideoCapture(video_path)

# Kiểm tra nếu video không mở được
if not cap.isOpened():
    print("Không thể mở video")
    exit()

# Đọc từng frame và chạy RT-DETR
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break  # Kết thúc video nếu không còn frame

    # Chạy RT-DETR trên frame hiện tại
    results = model(frame)

    # Vẽ bounding box lên frame
    annotated_frame = results[0].plot()

    # Resize frame để hiển thị tốt hơn (Ví dụ: 800x600)
    resized_frame = cv2.resize(annotated_frame, (800, 600))

    # Hiển thị video real-time
    cv2.imshow("RT-DETR Detection", resized_frame)

    # Nhấn 'q' để thoát
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Giải phóng bộ nhớ
cap.release()
cv2.destroyAllWindows()
```
